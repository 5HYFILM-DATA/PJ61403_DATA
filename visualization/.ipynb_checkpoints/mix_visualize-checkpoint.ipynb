```python
ad_dir = '/Users/waranthornchansawang/Desktop/RASH_DATA/mix_data/AtopicDermatitis'
pv_dir = '/Users/waranthornchansawang/Desktop/RASH_DATA/mix_data/PsoriasisVulgaris'
sk_dir = '/Users/waranthornchansawang/Desktop/RASH_DATA/mix_data/SeborrhoeicKeratosis'
vt_dir = '/Users/waranthornchansawang/Desktop/RASH_DATA/mix_data/Vitiligo'
```

```python
num_ad = len(os.listdir(ad_dir))
num_pv = len(os.listdir(pv_dir))
num_sk = len(os.listdir(sk_dir))
num_vt = len(os.listdir(vt_dir))

num_other = num_pv + num_sk + num_vt
```

```python
classes_mix_m1 = ['AD', 'Other']
classes_mix_m2 = ['AD', 'PV', 'SK', 'VT']

num_mix_per_classes_m1 = [num_ad, num_other]
num_mix_per_classes_m2 = [num_ad, num_pv, num_sk, num_vt]

num_mix_classes_m1 = np.arange(len(classes_mix_m1))
num_mix_classes_m2 = np.arange(len(classes_mix_m2))
```

```python
for a, b in (classes_mix_m1, num_mix_per_classes_m1):
    print(a, b)
    
print('')
    
for a, b, c, d in (classes_mix_m2, num_mix_per_classes_m2):
    print(a, b, c, d)
```

```python
plt.figure(figsize=(12, 4))
plt.bar(classes_mix_m1, num_mix_per_classes_m1)
plt.xticks(num_mix_classes_m1, classes_mix_m1, fontsize=15)
plt.xlabel('Classes')
plt.ylabel('Number of Picture')
plt.title('AD VS Other')
plt.show()

print('')

plt.figure(figsize=(12, 4))
plt.bar(classes_mix_m2, num_mix_per_classes_m2)
plt.xticks(num_mix_classes_m2, classes_mix_m2, fontsize=15)
plt.xlabel('Classes')
plt.ylabel('Number of Picture')
plt.title('AD VS Subclasses')
plt.show()
```
