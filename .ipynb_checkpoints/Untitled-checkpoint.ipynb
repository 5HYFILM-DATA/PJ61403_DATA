```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os
import cv2
import glob
```

```python
diry = '/Users/waranthornchansawang/Desktop/PJ61403_DATA/crop_classes_data/SeborrhoeicKeratosis/*.jpg'
```

```python
def plot_all_img_dir(directory):
    all_img_list = []
    for img in glob.glob(directory):
        img_array = cv2.imread(img)
        all_img_list.append(img_array)
    plt.figure(figsize=(10,10))
    
    columns = 5
    for i, image in enumerate(all_img_list):
        plt.subplot(len(all_img_list) / columns + 1, columns, i + 1)
        plt.imshow(image)
```

```python
plot_all_img_dir(diry)
```
