# Augmentation

```python
import keras
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(15,15))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
```

```python
path = 'Organ/arm'
```

```python
img_size = 224
```

fill_mode: One of {"constant", "nearest", "reflect" or "wrap"}. Default is 'nearest'. Points outside the boundaries of the input are filled according to the given mode:
'constant': kkkkkkkk|abcd|kkkkkkkk (cval=k)
'nearest': aaaaaaaa|abcd|dddddddd
'reflect': abcddcba|abcd|dcbaabcd
'wrap': abcdabcd|abcd|abcdabcd

```python
img_aug = ImageDataGenerator(rescale=1./255,
                            
                            rotation_range=360,
                            fill_mode='wrap') 
```

```python
img_gen = img_aug.flow_from_directory(directory=path,
                                     target_size=(224,224))
```

```python
img_list = [img_gen[0][0][0] for i in range(5)]
```

```python
plotImages(img_list)
```
